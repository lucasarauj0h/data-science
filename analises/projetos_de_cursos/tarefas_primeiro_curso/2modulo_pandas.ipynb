```
# O metodo "value_counts" faz a contagem dos valores presentes na lista e cria uma lista ordenada (do maior para o menor)
# Dos numeros que mais apareceram.
# O 'index[0]' seleciona apenas o indice 0 (nesse caso, o que mais apareceu, a moda) e atribui a var moda
moda = dsa_df['Quantidade'].value_counts().index[0]
moda
```

```
# Na coluna 'quantidade' use o metodo fillna (fill = preencha ; na "no atribuido") com o valor [value = moda] -> moda
# Inplace=True -> salve a alterao no proprio dataframe
dsa_df['Quantidade'].fillna(value=moda,inplace=True)
```

```
from sklearn.datasets import load_iris
data = load_iris()
```

```
import pandas as pd
df = pd.DataFrame(data['data'], columns = data['feature_names'])
df['species'] = data['target']
df.head()
```

```
df.plot()
```

```
# Que tal um scatter plot com duas variveis?
df.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)')
# Disperso de duas variaveis, que mostram a relao das duas variaveis
```

```
# E mesmo grficos mais complexos, como um grfico de rea, pode ser criado:
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']
df[columns].plot.area()
```

```
# Calculamos a mdia das colunas agrupando pela coluna species e criamos um grfico de barras com o resultado
df.groupby('species').mean().plot.bar()
```

```
# Ou ento, fazemos a contagem de classes da coluna species e plotamos em um grfico de pizza
df.groupby('species').count().plot.pie(y = 'sepal length (cm)')
```

```
# Grfico KDE (Kernel Density Function) para cada varivel do dataframe
df.plot.kde(subplots = True, figsize = (8,8))
```

```
# Boxplot de cada varivel numrica
columns = ['sepal length (cm)', 'petal length (cm)', 'petal width (cm)', 'sepal width (cm)']
df[columns].plot.box(figsize = (8,8))
```
