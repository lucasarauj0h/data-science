```
# Conecta no banco de dados
con = sqlite3.connect('cap12_dsa.db')
```

```
# Abre um cursor para percorrer os dados no banco de dados
cursor = con.cursor()
```

```
query = 'SELECT * FROM tb_vendas_dsa'
```

```
cursor.execute(query)
```

```
dados = cursor.fetchall()
dados
```

```
df = pd.DataFrame(dados, columns = ['ID_Pedido', 'ID_Cliente', 'Nome_Produto', 'Valor_Unitario', 'Unidades_Vendidas', 'Custo'])
```

```
df.head() # Data frame criado, vamos fechar o banco de dados.
```

```
cursor.close()
con.close()
```

```
media_unidades_vendidas = df['Unidades_Vendidas'].mean()
```

```
type(media_unidades_vendidas)
```

```
print(media_unidades_vendidas)
```

```
media_unidades_vendidas_por_produtos = df.groupby('Nome_Produto')['Unidades_Vendidas'].mean()
# Entre parenteses, parametro da função é o nome da coluna que desejo agrupar, em colchetes a coluna que aplicarei a média
```

```
media_unidades_vendidas_por_produtos
```

```
# Realizando um fateamento da minha tabela!
# 1º Filtro (Clausula WHERE)
# 2º Agrupamento (GROUP BY)
# 3º Calculo a média das colunas
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto')['Unidades_Vendidas'].mean()
df.head()
```

```
# Realizando um fateamento da minha tabela!
# 1º Filtro (Clausula WHERE)
# 2º Agrupamento (GROUP BY)
# 3º Calculo a média das colunas utilizando a função lambda
# 3-> nesta função lambda, eu atribuo a x[''] a coluna onde a media de unidades vendidas é > 10 e se for, ele atribui a x como verdadeiro e valida o filtro
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto').filter(lambda x: x['Unidades_Vendidas'].mean()>10)
```

```
# Realizando um fateamento da minha tabela!
# 1º Filtro (Clausula WHERE)
# 2º Agrupamento (GROUP BY)
# 3º Calculo a média das colunas utilizando a função lambda
# 3-> nesta função lambda, eu atribuo a x[''] a coluna onde a media de unidades vendidas é > 10 e se for, ele atribui a x como verdadeiro e valida o filtro
df[df['Valor_Unitario'] > 199].groupby('Nome_Produto').filter(lambda x: x['Unidades_Vendidas'].mean()>10).groupby('Nome_Produto')['Unidades_Vendidas'].mean()
```
