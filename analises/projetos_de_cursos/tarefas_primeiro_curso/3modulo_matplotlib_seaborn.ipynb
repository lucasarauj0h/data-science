```
# Grfico de linha

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0, 0, 0.8, 0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e ttulo
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Grfico de Linha');
```

```
# Grficos de linha com 2 figuras

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Cria os eixos
axes1 = fig.add_axes([0.3, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.4, 0.5, 0.4, 0.3]) # eixos da figura secundria

# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')

# Figura secundria
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundria');
```

```
# Grficos de linha em Paralelo

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Divide a rea de plotagem em dois subplots
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Ttulo')

# Ajusta o layout
fig.tight_layout()
```

```
# Grficos de linha com diferentes escalas

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize = (10,4))

# Cria o plot1
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padro")

# Cria o plot2
axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)");
```

```
# Grid

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize = (10,3))

# Grid padro
axes[0].plot(x, x**2, x, x**3, lw = 2)
axes[0].grid(True)

# Grid customizado
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```
# Diferentes estilos de Plots

# Dados
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize = (12,3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "black")
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n ** 2, lw = 2, color = "blue")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n ** 2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x ** 2, x ** 3, alpha = 0.5, color = "green");
axes[3].set_title("fill_between");
```

```
# Dados
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

# Funo para um mapa de cores
def ColorMap(phi_m, phi_p):
    return ( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

# Mais dados
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```
# Cria a figura
fig = plt.figure(figsize = (14,6))

# Adiciona o subplot 1 com projeo 3d
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Adiciona o subplot 2 com projeo 3d
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink=0.5)
```

```
# Wire frame

# Cria a figura
fig = plt.figure(figsize=(8,6))

# Subplot
ax = fig.add_subplot(1, 1, 1, projection = '3d')

# Wire frame
p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)
```

```
# Valores randmicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

# Cria os dados para a varivel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'No Fumante'})

# Revisar essa, fim da aula, modulo 11.
```

```
dados.shape
```

```
dados.head()
```

```
# Style
sea.set(style = "ticks") # Definindo o estilo para o seaborn

# lmplot
sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante', # Faa a colorao a partir da var / coluna 'flag_fumante'
           palette = ['tab:blue', 'tab:orange'], # paleta de cores
           height = 7)

# Labels e ttulo
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relao Entre Altura e Peso de Fumantes e No Fumantes')

# Remove as bordas
sea.despine()

# Show
plt.show()
```
