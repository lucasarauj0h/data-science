## Aprendizado Não Supervisionado

O aprendizado não supervisionado é uma abordagem de aprendizado de máquina (machine learning) onde os algoritmos são treinados em conjuntos de dados que não possuem rótulos ou categorias previamente definidas. O objetivo é encontrar padrões ou estruturas intrínsecas nos dados sem a necessidade de supervisão externa.

Isso pode ser feito por meio de diversas técnicas como agrupamento, redução de dimensionalidade e regras de associação.

Alguns exemplos de aplicação:

- **Segmentação de Clientes em um Supermercado:** Um supermercado pode usar aprendizado não supervisionado para agrupar os clientes com base em seus padrões de compra. Isso pode revelar diferentes tipos de clientes, como os que preferem produtos orgânicos, os que compram principalmente produtos de limpeza ou os que são mais propensos a comprar produtos promocionais.

- **Análise de Redes Sociais:** Em redes sociais, o aprendizado não supervisionado pode ser usado para agrupar os usuários com base em seus interesses, interações ou comportamentos. Isso pode ajudar a identificar comunidades de interesse ou detectar perfis falsos.

- **Classificação Automática de E-mails:** Um algoritmo de aprendizado não supervisionado pode analisar o conteúdo e os padrões de e-mails para agrupá-los em diferentes categorias, como "trabalho", "pessoal" ou "spam", sem a necessidade de rótulos prévios.

- **Detecção de Fraudes em Transações Financeiras:** Um banco pode usar aprendizado não supervisionado para identificar padrões suspeitos em transações financeiras e agrupar transações semelhantes para análise adicional, ajudando a detectar atividades fraudulentas.

- **Análise de Texto em Mídias Sociais:** Empresas de marketing podem usar aprendizado não supervisionado para analisar grandes volumes de texto em mídias sociais e identificar tópicos ou tendências emergentes entre os usuários.

#### K-means Clustering:

- **Aplicação:** Segmentação de clientes em um shopping com base em seus hábitos de compra.

- **Descrição:** O algoritmo K-means agrupa os clientes em diferentes segmentos com base em características semelhantes, como itens comprados, frequência de compra e valor gasto. Isso ajuda o shopping a personalizar ofertas e campanhas de marketing para cada segmento, melhorando a experiência do cliente e aumentando as vendas.


#### Análise de Componentes Principais (PCA):

- **Aplicação:** Redução de dimensionalidade em um conjunto de dados de imagens médicas.
- **Descrição:** O PCA é usado para reduzir a dimensionalidade de imagens médicas complexas, mantendo as características mais importantes. Isso permite uma representação mais compacta dos dados, facilitando a análise e o diagnóstico por parte dos médicos, além de reduzir os requisitos computacionais para o processamento das imagens.



#### DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

- **Aplicação:** Identificação de áreas de congestionamento de tráfego em uma cidade.

- **Descrição:** O DBSCAN identifica regiões densas de pontos em um conjunto de dados geográficos, como registros de GPS de veículos. Isso permite detectar áreas onde o tráfego é mais intenso, ajudando as autoridades de trânsito a tomar medidas para melhorar o fluxo de veículos e reduzir congestionamentos.



#### T-SNE (t-Distributed Stochastic Neighbor Embedding):

- **Aplicação:** Visualização de dados de palavras em um espaço semântico.

- **Descrição:** O T-SNE é usado para reduzir a dimensionalidade de dados de palavras em um espaço semântico, preservando as relações semânticas entre elas. Isso permite visualizar palavras similares próximas umas das outras e entender melhor a estrutura e as relações entre diferentes termos em um corpus de texto.



#### Regras de Associação (Apriori Algorithm):

- **Aplicação:** Análise de cestas de compras em um supermercado.
- **Descrição:** O algoritmo Apriori é usado para descobrir padrões de compra frequentes entre os itens comprados pelos clientes. Isso permite ao supermercado identificar produtos frequentemente comprados juntos e criar estratégias de marketing, como ofertas especiais e disposição de produtos em prateleiras, para aumentar as vendas e a satisfação do cliente.

### K-Means

O K-means é um algoritmo de aprendizado de máquina não supervisionado usado para agrupar dados em clusters. Ele agrupa os dados de forma que os pontos dentro de um cluster sejam mais semelhantes entre si do que com os pontos de outros clusters. O objetivo é encontrar K centróides, onde K representa o número de clusters pré-definidos pelo usuário.

O funcionamento básico do algoritmo K-means é o seguinte:

- **Inicialização:** Inicialize K centróides de forma aleatória no espaço de características.

- **Atribuição:** Para cada ponto de dado, calcule a distância até cada centróide e atribua o ponto ao cluster do centróide mais próximo.

- **Atualização:** Recalcule os centróides de cada cluster como a média de todos os pontos atribuídos a esse cluster.

- **Repetição:** Repita os passos 2 e 3 até que os centróides não mudem significativamente ou até que um critério de parada seja atingido (como um número máximo de iterações).


O K-means é amplamente utilizado em diversas áreas, como análise de dados, processamento de imagens, reconhecimento de padrões, entre outros. Ele é eficiente e fácil de implementar, mas é sensível à inicialização dos centróides e pode convergir para ótimos locais subótimos dependendo da configuração inicial. Além disso, o número de clusters K geralmente precisa ser especificado a priori, o que pode ser uma desvantagem em certos casos.


Alguns exemplos de aplicação do algoritmo K-means em diferentes áreas:

- **Segmentação de Mercado:** Uma empresa de varejo pode usar o K-means para segmentar clientes com base em padrões de compra, como frequência de compra, valor gasto e preferências de produtos. Isso permite direcionar campanhas de marketing específicas para cada segmento de clientes.

- **Análise de Imagem Médica:** Na área de saúde, o K-means pode ser aplicado para segmentar imagens médicas, como ressonâncias magnéticas ou tomografias computadorizadas, em diferentes regiões anatômicas com características semelhantes. Isso ajuda os médicos a identificar áreas de interesse e diagnosticar doenças.

- **Detecção de Anomalias em Redes de Computadores:** O K-means pode ser usado para agrupar padrões de tráfego de rede e identificar comportamentos anômalos, como ataques cibernéticos ou falhas de hardware. Isso permite que as equipes de segurança de TI detectem e respondam rapidamente a ameaças potenciais.

- **Classificação de Documentos Textuais:** Em processamento de linguagem natural, o K-means pode ser aplicado para agrupar documentos textuais em categorias semelhantes com base no conteúdo. Isso é útil para organizar grandes coleções de documentos, como artigos de notícias, e facilita a recuperação de informações.

- **Segmentação de Imagens em Visão Computacional:** Em aplicações de visão computacional, o K-means pode ser usado para segmentar imagens em regiões de interesse com base em cores ou texturas. Isso é útil para análise de imagens médicas, reconhecimento de objetos e detecção de padrões em imagens.

Esses são apenas alguns exemplos de como o algoritmo K-means pode ser aplicado em uma variedade de áreas para identificar padrões e estruturas nos dados. Sua versatilidade e simplicidade o tornam uma ferramenta poderosa em análise de dados e aprendizado de máquina

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Gerar dados de exemplo
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Aplicando o algoritmo k-means, mude o numero de kluster (numero de centros) para brincar e análisar
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# plotando os centroids
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('Clusters gerados pelo algoritmo K-means')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data[:, 2:4]  # Apenas as duas últimas características para simplificar a visualização

# Aplicar o algoritmo K-means
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# Calcular a distância euclidiana dos centros dos clusters ao ponto em azul (x=3, y=1.1)
point = np.array([3, 1.1])
distances = [np.linalg.norm(center - point) for center in kmeans.cluster_centers_]

# Plotar os clusters e os centróides
fig, ax = plt.subplots()

# Plotar os clusters
scatter = ax.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=30, alpha=0.3, cmap='viridis')

# Plotar o ponto em azul
ax.scatter(point[0], point[1], c='b', s=50, label='Ponto Azul')

# Plotar os centróides
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75, label='Centróides')

# Plotar a linha tracejada do centro do cluster ao ponto azul
for center in kmeans.cluster_centers_:
    ax.plot([center[0], point[0]], [center[1], point[1]], '--', color='black')
    dist = np.linalg.norm(center - point)
    if point[1] > center[1]:
        ax.text((center[0] + point[0]) / 2, (center[1] + point[1]) / 2, f'{dist:.2f}', fontsize=20, ha='center', va='bottom', color='red')
    else:
        ax.text((center[0] + point[0]) / 2, (center[1] + point[1]) / 2, f'{dist:.2f}', fontsize=20, ha='center', va='top', color='red')

# Plotar as linhas até o ponto em azul
ax.plot([kmeans.cluster_centers_[0][0], point[0]], [kmeans.cluster_centers_[0][1], point[1]], '--', color='black')
ax.plot([kmeans.cluster_centers_[1][0], point[0]], [kmeans.cluster_centers_[1][1], point[1]], '--', color='black')


plt.title('Clusters gerados pelo algoritmo K-means')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

#### Metricas de avaliação no aprendizado supervisionado

No aprendizado não supervisionado, como o K-means, por exemplo, pode ser um pouco mais desafiador avaliar o desempenho do modelo, já que não temos rótulos verdadeiros para comparar com as previsões do modelo. No entanto, existem algumas métricas que podem ser úteis na avaliação do desempenho de algoritmos de aprendizado não supervisionado:

- **Inércia (Within-cluster sum of squares):** A inércia mede a dispersão dos clusters. Quanto menor a inércia, mais compactos e bem definidos são os clusters. No entanto, a inércia sozinha pode não ser suficiente para determinar o número ideal de clusters.

- **Silhouette Score:** O Silhouette Score é uma medida de quão bem os objetos estão agrupados em seus clusters. Ele varia de -1 a 1, onde um valor alto indica que os objetos estão bem agrupados, um valor próximo de zero indica que os clusters se sobrepõem e um valor negativo indica que os objetos podem ter sido atribuídos ao cluster errado.

- **Davies-Bouldin Index:** O índice Davies-Bouldin é uma medida de dispersão entre os clusters. Quanto menor o valor do índice, melhor é a partição.

- **Calinski-Harabasz Index:** O índice Calinski-Harabasz é uma medida de dispersão entre e dentro dos clusters. Quanto maior o valor do índice, melhor é a partição.

- **Método do Cotovelo (Elbow Method):** O Método do Cotovelo é uma técnica gráfica para determinar o número ideal de clusters com base na inércia. O número ideal de clusters é geralmente encontrado no ponto em que a inércia começa a diminuir a uma taxa mais lenta, formando um "cotovelo" no gráfico.

Estas são algumas das métricas comuns utilizadas na avaliação de algoritmos de aprendizado não supervisionado. No entanto, é importante notar que a interpretação dessas métricas pode variar dependendo do contexto e da natureza dos dados. Experimentação e bom senso são muitas vezes necessários para determinar o desempenho adequado do modelo.

### Metodos de avaliação não supervisionado

O Índice Rand (Rand Index) e o Índice Rand Ajustado (Adjusted Rand Index) são métricas comuns usadas para avaliar a semelhança entre dois agrupamentos (ou partições) de um conjunto de dados. Essas métricas são frequentemente usadas em problemas de aprendizado não supervisionado, como clustering, para comparar a qualidade dos agrupamentos produzidos por diferentes algoritmos ou para avaliar a consistência dos agrupamentos em relação a rótulos conhecidos (quando disponíveis).

1. **Índice Rand (Rand Index):**

- O Índice Rand é uma medida simples de similaridade entre dois agrupamentos. Ele considera todos os pares de amostras no conjunto de dados e conta o número de pares que são atribuídos ao mesmo cluster em ambos os agrupamentos (pares concordantes) e o número de pares que são atribuídos a clusters diferentes em ambos os agrupamentos (pares discordantes).

- A fórmula do Índice Rand é: (a + b) / (a + b + c + d), onde:

  - a: número de pares concordantes (amostras no mesmo cluster em ambos os agrupamentos)

  - b: número de pares discordantes (amostras em clusters diferentes em ambos os agrupamentos)
  
  - c: número de pares onde as amostras estão no mesmo cluster no primeiro agrupamento, mas em clusters diferentes no segundo agrupamento

  - d: número de pares onde as amostras estão em clusters diferentes no primeiro agrupamento, mas no mesmo cluster no segundo agrupamento


- O Índice Rand varia de 0 a 1, onde 1 indica uma concordância perfeita entre os agrupamentos e 0 indica uma concordância aleatória.


2. **Índice Rand Ajustado (Adjusted Rand Index):**

- O Índice Rand Ajustado é uma versão ajustada do Índice Rand que leva em consideração o fato de que o Índice Rand tende a dar valores mais altos para conjuntos de dados com muitos clusters ou partições. O Índice Rand Ajustado corrige esse viés e fornece uma medida mais equilibrada da concordância entre os agrupamentos.


- A fórmula do Índice Rand Ajustado é uma correção do Índice Rand para aleatoriedade e variações no número de clusters. Ele é calculado subtraindo o valor esperado do Índice Rand sob a hipótese nula de aleatoriedade e normalizando pelo valor esperado do Índice Rand máximo. Assim, o Índice Rand Ajustado varia de -1 a 1, onde 1 indica uma concordância perfeita entre os agrupamentos, 0 indica uma concordância aleatória e valores negativos indicam uma concordância pior do que aleatória.


Em resumo, o Índice Rand e o Índice Rand Ajustado são métricas importantes para avaliar a qualidade dos agrupamentos em problemas de aprendizado não supervisionado, fornecendo uma medida quantitativa da concordância entre os agrupamentos.

```
valores_k = []
RI = []
ARI = []

for i in range(1,8):
  kmeans = KMeans(n_clusters=i, random_state=42).fit(X_iris)

  valores_k.append(i)
  RI.append(metrics.rand_score(y, kmeans.labels_))
  ARI.append(metrics.adjusted_rand_score(y, kmeans.labels_))
```

### Medida V

A medida V, também conhecida como Índice de Cramer ou Coeficiente de Cramer-V, é uma métrica estatística usada para avaliar a força da associação entre duas variáveis categóricas. Ela é uma generalização do coeficiente Phi (utilizado para tabelas de contingência 2x2) para tabelas de contingência maiores.

A medida V é calculada a partir da estatística qui-quadrado (chi-quadrado) associada a uma tabela de contingência, normalizada para estar entre 0 e 1. Quanto mais próximo de 1 for o valor de V, mais forte é a associação entre as variáveis.

A medida V varia de 0 a 1, onde:

0 indica nenhuma associação entre as variáveis,
1 indica uma associação perfeita entre as variáveis.
A medida V é comumente usada em análise de dados categóricos para avaliar a associação entre variáveis, como por exemplo, em tabelas de contingência de resultados de pesquisas ou em análise de dados demográficos. Ela fornece uma medida padronizada da força da associação, o que a torna útil para comparar a força de associação entre diferentes pares de variáveis.







image.png

```
import warnings

# Ignorar todos os warnings
warnings.filterwarnings("ignore")

X,y = load_iris(return_X_y=True, as_frame=True)
X_iris = X[['petal length (cm)','petal width (cm)']]

from sklearn.cluster import KMeans
from sklearn import metrics

valores_k = []
h = []
c = []
v = []

for i in range(1,25):
  kmeans = KMeans(n_clusters=i, random_state=42).fit(X_iris)

  valores_k.append(i)
  v.append(metrics.v_measure_score(y, kmeans.labels_))
  h.append(metrics.homogeneity_score(y, kmeans.labels_))
  c.append(metrics.completeness_score(y, kmeans.labels_))
```

```
fig, ax = plt.subplots()

ax.plot(valores_k, h, label='h')
ax.plot(valores_k, c, label='c')
ax.plot(valores_k, v, label='v')
ax.legend()
plt.show()
```

### Inércia

A inércia é uma métrica comum usada para avaliar a qualidade dos agrupamentos (clusters) em algoritmos de aprendizado não supervisionado, como o K-means. Ela representa a soma das distâncias quadradas de cada ponto ao centróide mais próximo, ou seja, mede o quão compactos e bem definidos são os clusters.

Quanto menor a inércia, mais compactos e bem definidos são os clusters, o que geralmente é desejável em problemas de agrupamento. No entanto, a inércia por si só não é suficiente para determinar o número ideal de clusters em um conjunto de dados, pois tende a diminuir à medida que o número de clusters aumenta.

Para avaliar o número ideal de clusters, é comum plotar um gráfico da inércia em função do número de clusters (também conhecido como "Método do Cotovelo"). Neste gráfico, a inércia geralmente diminui à medida que o número de clusters aumenta, mas a taxa de diminuição diminui gradualmente. O número ideal de clusters é geralmente escolhido no ponto onde a curva começa a diminuir mais lentamente, formando um "cotovelo" na curva.

No entanto, é importante observar que a interpretação do "cotovelo" pode ser subjetiva e depende do contexto do problema e da distribuição dos dados. Outras métricas e técnicas, como o índice de Silhueta ou a validação externa, podem ser usadas em conjunto com a inércia para selecionar o número ideal de clusters de forma mais robusta.

```
valores_k = []
inercias = []

for i in range(1,8):
  kmeans = KMeans(n_clusters=i, random_state=42).fit(X_iris)

  valores_k.append(i)
  inercias.append(kmeans.inertia_)
```

```
fig, ax = plt.subplots()

ax.plot(valores_k, inercias)

plt.show()
```

### Coeficiente Silhouette


O coeficiente de Silhueta (Silhouette Score) é uma métrica usada para avaliar a qualidade dos agrupamentos (clusters) em algoritmos de aprendizado não supervisionado, como o K-means. Ele fornece uma medida de quão bem os pontos de um cluster estão separados uns dos outros, em comparação com o quão bem eles estão separados dos pontos de outros clusters.

O coeficiente de Silhueta varia de -1 a 1, onde:

Um valor próximo de 1 indica que a amostra está longe dos clusters vizinhos e, portanto, bem ajustada ao seu próprio cluster.
Um valor próximo de 0 indica que a amostra está próxima da fronteira de decisão entre dois clusters.
Um valor próximo de -1 indica que a amostra pode ter sido atribuída ao cluster errado.

O coeficiente de Silhueta pode ser calculado para cada amostra individualmente e, em seguida, o valor médio é tomado como o coeficiente de Silhueta global do conjunto de dados. Um valor médio alto de coeficiente de Silhueta indica que os clusters estão bem separados, enquanto um valor médio baixo indica que os clusters podem estar se sobrepondo ou mal definidos.

O coeficiente de Silhueta é uma métrica útil para determinar o número ideal de clusters em um conjunto de dados, pois pode ser usado para comparar a qualidade dos agrupamentos para diferentes valores de k no K-means. Um valor máximo ou próximo de 1 do coeficiente de Silhueta geralmente indica um número ideal de clusters. No entanto, é importante considerar outros aspectos do problema e das características dos dados ao selecionar o número ideal de clusters.

### PCA (Principal Component Analysis)

image.png

PCA (Principal Component Analysis), ou Análise de Componentes Principais em português, é uma técnica de redução de dimensionalidade amplamente utilizada em aprendizado de máquina e análise de dados. Seu principal objetivo é reduzir a complexidade de conjuntos de dados de alta dimensionalidade, preservando ao máximo a estrutura e a variabilidade dos dados.

Em termos simples, o PCA busca encontrar um novo conjunto de variáveis, chamadas de componentes principais, que são combinações lineares das variáveis originais. Esses componentes principais são ordenados de forma que o primeiro capture a maior parte da variabilidade dos dados, o segundo capture a segunda maior parte e assim por diante. Dessa forma, podemos representar os dados em um espaço de dimensionalidade reduzida, mantendo a maior parte da informação contida nos dados originais.

Os passos básicos do PCA são os seguintes:

- Padronização dos dados: Os dados são centralizados na média e escalonados para ter variância unitária em cada dimensão.
- Cálculo da matriz de covariância ou matriz de correlação: Isso nos fornece informações sobre as relações lineares entre as variáveis originais.
- Decomposição da matriz de covariância: A matriz de covariância é decomposta em autovetores e autovalores.
- Seleção dos componentes principais: Os autovetores correspondentes aos maiores autovalores representam os componentes principais.
- Projeção dos dados nos novos eixos: Os dados originais são projetados nos novos eixos definidos pelos autovetores.

O PCA é amplamente utilizado para visualização de dados, redução de dimensionalidade, remoção de ruído e seleção de características. Ele é particularmente útil em casos onde há muitas variáveis correlacionadas ou redundantes, pois pode ajudar a identificar as direções de máxima variância nos dados.

Em resumo, o PCA é uma ferramenta poderosa para explorar e analisar conjuntos de dados complexos, permitindo uma compreensão mais clara das relações entre as variáveis e facilitando a interpretação dos resultados.

**Análise de Componentes Principais**

Técnica não supervisionada de redução de dimensionalidade através de combinações lineares dos recursos (features)  
Variáveis relacionadas (com relação entre si) são reduzidas a um número menor de componentes não relacionados      

- O primeiro componente principal tenta explicar o máximo da variância dos dados

  * O segundo tenta explicar o máximo da variância não explicada pelo primeiro
    * O terceiro, a variância não explicada pelo primeiro nem pelo segundo
      * O quarto...

Facilita a visualização em problemas de clusterização, viabiliza análises em casos com muitas variáveis, melhora os
modelos, etc.
