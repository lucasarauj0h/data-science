```
%%sql

CREATE TABLE pacientes (Pregnancies INT,
                        Glucose INT,
                        BloodPressure INT,
                        SkinThickness INT,
                        Insulin INT,
                        BMI DECIMAL(8, 2),
                        DiabetesPedigreeFunction DECIMAL(8, 2),
                        Age INT,
                        Outcome INT);
```

```
%%sql

SELECT * FROM pacientes
```

```
%%sql

# movendo os mesmos dados para uma nova tabela, where (apenas aqueles com 50+ anos)

INSERT INTO pacientes(Pregnancies,
                      Glucose,
                      BloodPressure,
                      SkinThickness,
                      Insulin,
                      BMI,
                      DiabetesPedigreeFunction,
                      Age,
                      Outcome)
SELECT Pregnancies,
       Glucose,
       BloodPressure,
       SkinThickness,
       Insulin,
       BMI,
       DiabetesPedigreeFunction,
       Age,
       Outcome
FROM diabetes WHERE Age > 50;
```

```
%%sql

SELECT * FROM pacientes
```

```
%%sql
# Criando a coluna perfil
ALTER TABLE pacientes
ADD Perfil VARCHAR(10);
```

```
%%sql
# Configurando a coluna perfil: fazendo um update na tabela 'pacientes', na coluna perfil eu defino com o'normal' qnd bmi < 30
UPDATE pacientes
SET Perfil = 'Normal'
WHERE BMI < 30;
```

```
%%sql
# Configurando a coluna perfil: fazendo um update na tabela 'pacientes', na coluna perfil eu defino com o 'obeso' qnd bmi > 30
UPDATE pacientes
SET Perfil = 'Obeso'
WHERE BMI > 30;
```
